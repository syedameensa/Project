# label encoder and standard scaler

```python
df.head()
```

```python
Cont = df[df.columns[df.dtypes != 'O']]
Cat = df[df.columns[df.dtypes == 'O']]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in Cat.columns:
    Cat[i] = le.fit_transform(Cat[i])
```

```python
Cat.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
Cont = pd.DataFrame(sc.fit_transform(Cont),columns = Cont.columns)
```

```python
Cont.tail()
```

```python
Cont.shape
```

```python
Cat.shape
```

```python
Cat.tail()
```

```python
df = pd.concat([Cat,Cont],axis = 1)
```

```python
df.drop([7042,0],inplace = True)
```

```python
df.head()
```

```python
df.SeniorCitizen = df.SeniorCitizen.astype(int)
df.Partner = df.Partner.astype(int)
df.Dependents = df.Dependents.astype(int)
df.MultipleLines = df.MultipleLines.astype(int)
df.InternetService = df.InternetService.astype(int)
df.OnlineSecurity = df.OnlineSecurity.astype(int)
df.OnlineBackup = df.OnlineBackup.astype(int)
df.DeviceProtection = df.DeviceProtection.astype(int)
df.TechSupport = df.TechSupport.astype(int)
df.StreamingTV = df.StreamingTV.astype(int)
df.StreamingMovies = df.StreamingMovies.astype(int)
df.PaperlessBilling = df.PaperlessBilling.astype(int)
df.PaymentMethod = df.PaymentMethod.astype(int)
df.Churn = df.Churn.astype(int)
```

```python
df.head()
```

```python
df.Churn.value_counts()
```

```python
from sklearn.utils import resample
```

```python
one = df[df.Churn == 1]
zero= df[df.Churn == 0]
```

```python
oversample = resample(one,replace = True,n_samples = len(zero),random_state = 8)
```

```python
df = pd.concat([oversample,zero])
```
